```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

df = pd.read_excel("data_kuesioner.xlsx")
pertanyaan = df.iloc[:, 1:]   # Q1â€“Q17
semua_jawaban = pertanyaan.values.flatten()
```

```python
counter = Counter(semua_jawaban)

plt.bar(counter.keys(), counter.values())
plt.title("Distribusi Skala Jawaban Keseluruhan")
plt.xlabel("Skala")
plt.ylabel("Jumlah")
plt.show()
```

```python
target = "S"
jumlah_per_pertanyaan = []

for col in pertanyaan.columns:
    jumlah = (pertanyaan[col] == target).sum()
    jumlah_per_pertanyaan.append(jumlah)

plt.bar(pertanyaan.columns, jumlah_per_pertanyaan)
plt.title("Jumlah Skala S per Pertanyaan")
plt.xlabel("Pertanyaan")
plt.ylabel("Jumlah")
plt.xticks(rotation=45)
plt.show()
```

```python
skor = {
    "SS": 6,
    "S": 5,
    "CS": 4,
    "CTS": 3,
    "TS": 2,
    "STS": 1
}
rata = pertanyaan.replace(skor).mean()

plt.bar(rata.index, rata.values)
plt.title("Rata-rata Skor per Pertanyaan")
plt.xticks(rotation=45)
plt.show()
```
